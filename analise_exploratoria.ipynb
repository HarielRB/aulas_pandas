```python
#importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
#criando dataframe
df = pd.read_excel("/content/AdventureWorks.xlsx")
```

```python
#visualizando a head
df.head()
```

```python
#quantidade de linhas e colunas
df.shape
```

```python
#tipos de dados
df.dtypes
```

```python
#receita total das vendas
df['Valor Venda'].sum()
```

```python
#custo 
df['custo'] = df['Custo Unitário'].mul(df['Quantidade'])
```

```python
#custo total
round(df['custo'].sum(), 2)
```

```python
#lucro
df['lucro'] = df['Valor Venda'] - df['custo']
```

```python
df.head(1)
```

```python
#lucro total
round(df['lucro'].sum(), 2)
```

```python
#analisando o tempo de demora para envio
df['tempo_envio'] = df['Data Envio'] - df['Data Venda']
```

```python
#extraindo apenas os dias
df['tempo_envio'] = (df['Data Envio'] - df['Data Venda']).dt.days
```

```python
df['tempo_envio'].dtypes
```

```python
#verificando media do tempo de envio por marca
df.groupby("Marca")["tempo_envio"].mean()
```

```python
#verificando se existem valores nulos
df.isnull().sum()
```

```python
#agrupando por ano e por marca para descobrir o luco
df.groupby([df["Data Venda"].dt.year, "Marca"])['lucro'].sum()
```

```python
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
#resetando os index
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])['lucro'].sum().reset_index()
lucro_ano
```

```python
#total de produtos vendidos
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)
```

```python
#Grafico total de produtos vendidos
df.groupby("Produto")["Quantidade"].sum().sort_values().plot.barh(title="Total Produtos Vendidos")
plt.xlabel('Total')
plt.ylabel("Produto");
plt.savefig("total_produtos_vendidos.png")
```

```python
df.groupby(df['Data Venda'].dt.year)["lucro"].sum().plot.bar(title="Lucro x Ano")
plt.xlabel('Ano')
plt.ylabel("Receita");
plt.savefig("total_receita_ano.png")
```

```python
#selecionando vendas de 2009
df_2009 = df[df['Data Venda'].dt.year == 2009]
```

```python
df_2009.head()
```

```python
df_2009.groupby(df_2009['Data Venda'].dt.month)['lucro'].sum().plot(title='Lucro X Mes')
plt.xlabel("Mês")
plt.ylabel("Lucro");
```

```python
df_2009.groupby("Marca")['lucro'].sum().plot.bar(title="Lucro X Marca")
plt.xlabel("Marca")
plt.ylabel("Lucro")
plt.xticks(rotation="horizontal")
```

```python
df_2009.groupby("Classe")['lucro'].sum().plot.bar(title="Lucro X Classe")
plt.xlabel("Classe")
plt.ylabel("Lucro")
plt.xticks(rotation="horizontal")
```
